```python
import pandas as pd
from src.paths import TRANSFORMED_DATA_DIR


df = pd.read_parquet(TRANSFORMED_DATA_DIR / "tabular_data.parquet")
df
```

```python
from datetime import datetime
from src.data_split import train_test_split # import func from py file

X_train, y_train, X_test, y_test = train_test_split(df, cutoff_date=datetime(2022, 6, 1, 0, 0, 0), target_column_name="target_rides_next_hour")

print(f"X_train: {X_train.shape}") # (examples-in-train, num-features)
print(f"y_train: {y_train.shape}") # (examples-in-train,) 1D beceause only one target
print(f"X_test: {X_test.shape}")   # (examples-in-test, num-features)
print(f"y_test: {y_test.shape}")   # (examples-in-test,) 1D beceause only one target
# examples-in-train != examples-in-test
 
```

```python
import numpy as np

# class for baseline-model to compare with more complex models in fututre
class BaselineModelPreviousHour:
    """
    Predicts actual deman observed in the last hour
    """
    # takes in training-features-targets, usually model adjusts parameters in this step, but there is nothing to adjust no parameters so it essentially does nothing
    def fit(self, X_train, y_train):
        pass
    
    # takes in test-feautres and returns the demand from preivous hour
    def predict(self, X_test):
        return X_test[f"rides_previous_1_hour"]
```

```python
model = BaselineModelPreviousHour()   # create baseline-model-obj
predictions = model.predict(X_test)   # generate predictions base on testing-features
predictions    # print vector of predictions
```

```python
# MAE for error function, this is regression problem
from sklearn.metrics import mean_absolute_error

test_mae = mean_absolute_error(y_test, predictions)  # built-in-func takes in ltesting-labels and predictions-vector, outputs mean-absolute-error
print(f"MAE: {test_mae=:.4f}")
```
